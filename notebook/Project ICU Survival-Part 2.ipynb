# Objective for Part 2

We will perform EDA on the data using the following techniques:
- Univariate Analysis
- Bivariate Analysis

```python
# Step 1: Import the libraries
#basic libs
import pandas as pd

#visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Read your semi-cleaned CSV as DataFrame
df = pd.read_csv(r"data/data_part1.csv", index_col=0)
```

```python
print(df["hospital_death"].value_counts())
```

```python
# Step 3: Plot a barplot for "hospital_death"
df["hospital_death"].value_counts().plot(kind="bar")
plt.show()
```

```python
#Matplotlib barplot for "hospital_death"
death = df["hospital_death"].value_counts().reset_index()
x=death.iloc[:,0]
y=death.iloc[:,1]
plt.bar(x=x,height=y)
plt.show()
```

```python
#Seaborn barplot for "hospital_death"
sns.barplot(x=x, y=y)
plt.show()
```

```python
#Seaborn countplot for "hospital_death"
sns.countplot(x="hospital_death",data=df)
plt.show()
```

```python
# Step 4: Plot a histogram with "age"
df.hist(column=["age"], bins=10)
plt.show()
```

```python
# Step 5: Plot a histogram with "bmi"
df.hist(column=["bmi"], bins=[0,10,20,30,40,50,60,70])
plt.show()
```

```python
# Step 6: Plot a barplot with 'elective_surgery'
sns.countplot(x="elective_surgery",data=df)
plt.show()
```

```python
# Step 7: Plot a barplot with 'ethnicity'
sns.countplot(y="ethnicity",data=df)
plt.show()
```

```python
# Step 8: Plot a barplot with "gender"
sns.countplot(x="gender",data=df)
plt.show()
```

```python
# Step 9: Plot a histogram with 'height'
df.hist(column=["height"], bins=[120,130,140,150,160,170,180,190])
plt.show()
```

```python
# Step 10: Plot a histogram with 'pre_icu_los_days'
df.hist(column=["pre_icu_los_days"], bins = [0,5,10,15,20,25,30])
plt.show()
```

```python
# Step 11: Plot age vs hospital_death with a boxplot
sns.boxplot(x=df["hospital_death"], y=df["age"] )
plt.show()
```

```python
# Step 12: Plot bmi vs hospital_death with boxplot
sns.boxplot(x=df["hospital_death"], y=df["bmi"])
plt.show()
```

```python
# Step 13: Plot a barplot
sns.boxplot(x=df["hospital_death"], y=df["pre_icu_los_days"], showfliers = False)
plt.show()
```

```python
# Step 14: Plot a grouped countplot with seaborn for gender vs hospital death
sns.countplot(x="gender",hue="hospital_death",data=df)
plt.show()
```

```python
# Step 15: Plot a boxplot with weight vs hospital death
sns.boxplot(x="hospital_death", y="weight", data=df)
plt.show()
```
